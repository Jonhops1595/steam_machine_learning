Importing Clean Data

```python
import pandas as pd
import datetime

pd.set_option('display.max_column', None)

game_data = pd.read_csv('cleaned_game_data.csv', index_col = 0)

#Converting release_date to datetime
game_data['release_date'] = pd.to_datetime(game_data['release_date'])
game_data
```

```python
#Converting True/False back to 1/0s
game_data = game_data.replace({True:1.0, False:0.0}) #Convert to True/False
```

Trees for free_to_play

```python
from sklearn import tree
#Cleaning data and selecting variables
tree_data=game_data.dropna(subset='free_to_play')
tree_data= tree_data.drop(['app_id',
                                'name',
                                'genre_Free to Play',
                                'type',
                                'release_date',
                                'publisher_1',
                                'publisher_2',
                                'publisher_3',
                                'developer_1',
                                'developer_2',
                                'developer_3',
                                'first_availible_price',
                                'last_availible_price',
                                'lowest_price',
                                'highest_price'],
                               axis=1)
#Replacing genre nulls with 0s
tree_data.fillna(0, inplace = True)
tree_data
```

```python
#Splitting Dataset
from sklearn.model_selection import train_test_split

train, test = train_test_split(tree_data, test_size=.2,random_state=95)
```

```python
#Creating Test Data 
X = train.drop(['free_to_play'],axis=1)
Y = train['free_to_play']
```

```python
#Fitting Tree
clf = tree.DecisionTreeClassifier(criterion='entropy',max_depth = 3)
clf = clf.fit(X, Y)
```

```python
#Visualizing Tree
import pydotplus 
from IPython.display import Image

dot_data = tree.export_graphviz(clf, feature_names=X.columns, class_names=['free_to_play','not_free_to_play'], filled=True, 
                                out_file=None) 
graph = pydotplus.graph_from_dot_data(dot_data) 
Image(graph.create_png())

```

```python
#Testing
X_test = test.drop(['free_to_play'],axis=1)
predictions = clf.predict(X_test)
```

```python
#Accuracy Test
from sklearn.metrics import accuracy_score
Y_test = test['free_to_play']
print(accuracy_score(Y_test,predictions))
```
